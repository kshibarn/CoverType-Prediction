# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,confusion_matrix, accuracy_score

sns.set_style('darkgrid')
```

```python
ran_forest= pd.read_csv("covtype.csv")
```

## Selecting Useful Columns

```python
ran_forest = ran_forest[['Elevation', 'Aspect', 'Slope', 'Horizontal_Distance_To_Hydrology','Vertical_Distance_To_Hydrology', 
                   'Horizontal_Distance_To_Roadways', 'Hillshade_9am', 'Hillshade_Noon', 'Hillshade_3pm',
                   'Horizontal_Distance_To_Fire_Points', 'Cover_Type']]
```

# Train Test Split
#### Now let's split the data into a training set and a testing set. We will train out model on the training set and then use the test set to evaluate the model.

```python
X = ran_forest.drop(["Cover_Type"], axis=1)
y = ran_forest["Cover_Type"]
```

```python
X.head()
```

```python
y=pd.DataFrame(y)
y.head()
```

```python
X.shape
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.25,random_state=42)
```

```python
len(X_test)
```

# Creating and Training the Model

```python
rf = RandomForestClassifier(n_estimators=200, class_weight='balanced', n_jobs=2, random_state=42, bootstrap=True)
rf.fit(X_train,y_train)
print("Training is Completed")
```

# Predictions from our Model

#### Let's grab predictions off our test set and see how well it did!

```python
y_predict = rf.predict(X_test)
```

```python
len(y_predict)
```

```python
print("Training Score:",rf.score(X_train,y_train))
print("Test Score:",rf.score(X_test,y_test))
```

```python
print("accuracy_score: " + str(accuracy_score(y_test, y_predict)))
print("\n")
print("classification_report:\n",classification_report(y_test,y_predict))
print("\n")
print("confusion_matrix:\n ",confusion_matrix(y_test,y_predict))
```

```python
cf_matrix = confusion_matrix(y_test,y_predict)
plt.figure(figsize=(7,5))
sns.heatmap(cf_matrix, annot=True, linewidths = 1.0, cmap = 'crest', linecolor = 'black')
```
